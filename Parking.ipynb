```python
#chosen zip code: 10025, and grab all rows with 'Incident Zip' = 10025
_10025 = _311.loc[_311['Incident Zip'] == 10025]

#Grab all rows that 'Complaint Type' is 'Illegal Parking'
illegal_park_10025 = _10025.loc[_10025['Complaint Type'] == 'Illegal Parking']
```

```python
#divide the total number of parking incidents in my ZIP by the total number of all incidents in my ZIP
local_parking_incident_rate = len(illegal_park_10025)/len(_10025)
local_parking_incident_rate
```

```python
#divide the total number of parking incidents in NYC by the total number of all incidents in NYC
all_parking_incidents =  _311.loc[_311['Complaint Type'] == 'Illegal Parking']
global_parking_incident_rate = len(all_parking_incidents)/ len(_311)
global_parking_incident_rate
```

```python
higher_parking_proportion = False

if local_parking_incident_rate > global_parking_incident_rate:
    higher_parking_proportion = True
```
